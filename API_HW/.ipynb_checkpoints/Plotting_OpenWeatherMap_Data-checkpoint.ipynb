### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

```python
# Dependencies
import matplotlib.pyplot as plt
%matplotlib notebook
import pandas as pd

# Import saved OpenWeatherMap cities data
file = 'OpenWeatherMap_Data_538_Cities.csv'
weather_df = pd.read_csv(file)

# Check weather_df
weather_df.head()
```

```python
# Clean up weather_df by dropping Unnamed:0 column
weather_df.drop(columns=['Unnamed: 0'], inplace=True)

# Check cleaned up weather_df
weather_df.head()
```

#### Latitude vs. Temperature Plot

```python
# Plot cities' latitudes against the temperatures there

# Create x and y label ticks
xticks = [-80,-60, -40, -20, 0, 20, 40, 60, 80, 100]
yticks = [-100, -50, 0, 50, 100, 150]

# Plot figure with pandas
lat_temp_fig = weather_df.plot('Latitude', 'Temperature', 
                kind='scatter',
                title='City Latitude vs Temperature (F) 1/5/2019',
                grid=True,
                xticks=xticks, 
                yticks=yticks,)

# Modify axis labels with matplotlib
plt.xlabel("City Latitude")
plt.ylabel("City Temperature (F)")

# Save pie chart and include note in saved figure
plt.savefig("Latitude_v_Temperature.png", bbox_inches='tight')
```
